```python
import pandas as pd
import requests
import logging
import time
import re
import string
from timeit import default_timer as timer
from glob import glob
from shapely.geometry import Polygon, Point
import geopandas as gpd
```

# Prepare Data

```python
path_to_data='../../data/'
```

```python
print('Import Twitter POIS:')
start = timer()

files = glob('../../../twitter/data/locations/pois/*.snappy.parquet')

pois = pd.DataFrame()
for file in files:
    pois = pd.concat([pois,pd.read_parquet(file)])

pois = pois.sort_values(by='n_obs',ascending=False).reset_index(drop=True)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
pois.head()
```

```python
print('# Twitter Places',pois.shape[0])

# Select POIs
pois = pois[(pois['place_type']=='poi')].copy()
print('# Twitter Pois',pois.shape[0])

# Select POIs in Mexico
pois = pois[(pois['country_code']=='MX')].copy()
print('# Twitter Pois in Mexico',pois.shape[0])

pois.dropna(inplace=True)

# Geometry
pois['coordinates'] = pois['coordinates'].apply(lambda x:Polygon([list(y) for y in x[0]]))

# Collapse Null Areas to Single Point
if not pois['coordinates'].apply(lambda x:x.area).sum():
    pois['coordinates'] = pois['coordinates'].apply(lambda x:x.centroid)
else:
    print('Check Geometry')
    
pois['longitude'] = pois['coordinates'].apply(lambda x:x.x)
pois['latitude'] = pois['coordinates'].apply(lambda x:x.y)

pois = pois.sort_values(by='n_obs',ascending=False).reset_index(drop=True)

pois.drop(['type','country','country_code','name','place_type','url','coordinates'],1,errors='ignore',inplace=True)
```

```python
pois.head()
```

# Params

```python
# input_    = "Chilis Reforma 222, Cuauhtémoc"
inputtype = "textquery"
fields    = "formatted_address,geometry,icon,name,permanently_closed,photos,place_id,plus_code,types,opening_hours,price_level,rating,user_ratings_total"
radius    = 500
# latitude  = 19.429077
# longitude = -99.161886
with open(path_to_data+'keys/google/API_KEY','r') as f:
    api_key=eval(f.readline())
    
# Backoff time sets how many minutes to wait between google pings when your API limit is hit
BACKOFF_TIME = 30

output_file = 'pois/pois-twitter-mexico-geocoded-google-place-api.pkl'
```

```python
logger = logging.getLogger("root")
logger.setLevel(logging.DEBUG)

# create console handler
ch = logging.StreamHandler()
ch.setLevel(logging.DEBUG)
logger.addHandler(ch)
```

# API QUERY

```python
def google_place_search(input_, inputtype, fields, radius, latitude, longitude, api_key):
    
    # Set up place url
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"+\
    "input="+input_+"&inputtype="+inputtype+"&fields="+fields+"&"+\
    "locationbias=circle:"+str(radius)+"@"+str(latitude)+","+str(longitude)+"&"+\
    "key="+api_key

    # Ping google for the reuslts:
    results = requests.get(url)
    
    # Results will be in JSON format - convert to dict using requests functionality
    return results.json()
```

```python
# Create a list to hold results
results = []

# Go through each place in turn
for idx, poi in pois.iterrows():
    
    # While the place geocoding is not finished:
    geocoded = False
    
    while geocoded is not True:
        
        # Geocode the place with google
        try:
            
            geocode_result = google_place_search(poi['full_name'], 
                                                 inputtype, 
                                                 fields, 
                                                 radius, 
                                                 poi['latitude'], 
                                                 poi['longitude'], 
                                                 api_key)
            
            geocode_result['id'] = poi['id']
            
        except Exception as e:
            
            logger.exception(e)
            
            logger.error("Major error with {}".format(poi['full_name']))
            
            logger.error("Skipping!")
            
            geocoded = True
            
        # If we're over the API limit, backoff for a while and try again later.
        if geocode_result['status'] == 'OVER_QUERY_LIMIT':
            
            logger.info("Hit Query Limit! Backing off for a bit.")
            
            time.sleep(BACKOFF_TIME * 60) # sleep for 30 minutes
            
            geocoded = False
            
        else:
            
            # If we're ok with API use, save the results
            # Note that the results might be empty / non-ok - log this
            if geocode_result['status'] != 'OK':
                
                logger.warning("Error geocoding {}: {}".format(poi['full_name'], geocode_result['status']))
                
            logger.debug("Geocoded: {}: {}".format(poi['full_name'], geocode_result['status']))
            
            results.append(geocode_result)           
            
            geocoded = True

    # Print status every 100 places
    if len(results) % 100 == 0:
        
        logger.info("Completed {} of {} places".format(len(results), len(pois)))
            
    # Every 1000 places, save progress to file (in case of a failure)
    if len(results) % 1000 == 0:
        
        pd.DataFrame(results).to_pickle("{}_bak".format(output_file))

# All done
logger.info("Finished geocoding all places")
```

```python
print('SAVE:')

# Write the full results to csv using the pandas library.
pd.DataFrame(results).to_pickle(path_to_data+output_file)

print('DONE!')
```

```python
print('Import Raw Geocodes')
geocodes = pd.read_pickle(output_file)
print('# Geocoded POIS:', geocodes.shape[0])
```

```python
geocodes.head()
```

```python
print('Merge POIS and their Geocodes')

pois = pd.merge(pois,geocodes,on='id')
print('# Geocoded POIS:', pois.shape[0])

pois = pois[pois['status']=='OK'].drop(['error_message','status'],1).reset_index(drop=True)
print('# Geocoded POIS:', pois.shape[0])
```

```python
pois['northeast'] = pois['candidates'].apply(lambda x:list(x[0]['geometry']['viewport']['northeast'].values()))
pois['southwest'] = pois['candidates'].apply(lambda x:list(x[0]['geometry']['viewport']['southwest'].values()))
pois['name_google'] = pois['candidates'].apply(lambda x:x[0].get('name',None))
pois['address'] = pois['candidates'].apply(lambda x:x[0].get('formatted_address',None))
pois['types'] = pois['candidates'].apply(lambda x:x[0].get('types',None))

pois.dropna(inplace=True)
print('# Geocoded POIS:', pois.shape[0])

pois['type'] = pois['types'].apply(lambda x:x[0])
```
