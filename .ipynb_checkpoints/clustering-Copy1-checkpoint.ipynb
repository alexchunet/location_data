In more general case, allow merges between urban areas if distance below a threshold.

If blocks are always separated by streets of urban amenities then the average distance between blocks that are allowed to merged will be much less than the one between blocks which are parts of two cities which are far away to one another. So we would probably get the same results using a statistical criterion.

```python
from timeit import default_timer as timer
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.ops import cascaded_union, unary_union, polygonize
from shapely.geometry import MultiPolygon, Polygon
import multiprocessing as mp
import matplotlib.pyplot as plt
```

```python
print('Import Census Blocks')
start = timer()

blocks = gpd.read_file('../data/blocks-mexico.geojson')
print('# Blocks:', blocks.shape[0])

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
blocks.set_index('CVEGEO',inplace=True)
```

```python
blocks.head()
```

```python
empty_blocks = blocks.loc[blocks['POB1']==0].index.tolist()
print('Empty Blocks:', len(empty_blocks))
```

# Fix Invalid Blocks

```python
def make_valid(invalid_polygon):
    
    # Take exterior
    pe = invalid_polygon.exterior

    # Intersect with itself
    mls = pe.intersection(pe)

    # Create a polygon from multiline string
    polygons = polygonize(mls)

    # Aggregate into Multi-poloyogns
    return MultiPolygon(polygons)
```

```python
print('Fix Invalid Blocks')
start = timer()

invalid_blocks = blocks[-blocks.geometry.is_valid].copy()
print('# Invalid Blocks:', invalid_blocks.shape[0])

valid_blocks = invalid_blocks.copy()
valid_blocks.geometry = valid_blocks.geometry.apply(make_valid)
blocks.loc[invalid_blocks.index,'geometry'] = valid_blocks['geometry'].copy()

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

# Agglomerative Clustering of Children Blocks

```python
print('Extract Nested Blocks')
start = timer()

def nested_shapes(df,index_col='CVEGEO'):
    
    # Find Shapes that are Contained in one Another
    nested_shapes = gpd.sjoin(
    df[['geometry']].reset_index(),
    df[['geometry']].reset_index(),
    op='contains',
    lsuffix='parent',
    rsuffix='child').drop(
    ['index_child','geometry'],1)
    
    # Remove Original Shapes
    nested_shapes = nested_shapes[
    nested_shapes[(index_col+'_parent')]!=\
    nested_shapes[(index_col+'_child')]].sort_values(
    by=[index_col+'_parent',index_col+'_child']).reset_index(drop=True)
    
    return nested_shapes

nested_blocks = nested_shapes(blocks)
print('Nested Blocks:', nested_blocks.shape[0])
    
end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
nested_blocks.head()
```

```python
def cluster_shapes(clusters, data=blocks, pop_col='POB1', pop_threshold=100):
    
    # Initialize clusters populations
    pops = [data.loc[cluster,pop_col].sum() for cluster in clusters]

    # Initialize clusters geometry
    geoms = [unary_union(data.loc[cluster,'geometry']) for cluster in clusters]
        
    # Stop if there is only one cluster left
    while len(clusters) > 1:

        # Create random vector to break ties
        np.random.seed()
        pops_tiebreak = np.random.random(len(pops))
        
        # Find the position of the cluster with the smallest population
        pos_smallest = np.lexsort((pops_tiebreak, pops))[0]

        # Stop if the smallest cluster is above the threshold
        if pops[pos_smallest] >= pop_threshold:
            break

        # Compute distances to the smallest cluster
        dists = [geom.distance(geoms[pos_smallest]) for geom in geoms]
        
        # Create random vector to break ties
        np.random.seed()
        dists_tiebreak = np.random.random(len(dists))
        
        # Find the position of the smallest cluster's closest neighboring cluster
        pos_dists   = np.lexsort((dists_tiebreak, dists))
        pos_closest = [pos for pos in pos_dists if pos != pos_smallest][0]
        
        # Create a new cluster merging the smallest cluster and its closest neighbor
        new_cluster = clusters[pos_smallest] + clusters[pos_closest]
        new_pop     = pops[pos_smallest] + pops[pos_closest]
        new_geom    = unary_union([geoms[pos_smallest], geoms[pos_closest]])
        
        # Remove smallest cluster and its closest neighbor
        clusters = [cluster for pos_cluster,cluster in enumerate(clusters) if pos_cluster not in [pos_smallest,pos_closest]]
        pops     = [pop for pos_pop,pop in enumerate(pops) if pos_pop not in [pos_smallest,pos_closest]]
        geoms    = [geom for pos_geom,geom in enumerate(geoms) if pos_geom not in [pos_smallest,pos_closest]]

        # Append the new cluster and its population
        clusters.append(new_cluster)
        pops.append(new_pop)
        geoms.append(new_geom)
        
    return clusters
```

```python
start = timer()
print('Cluster Children')

parent2children = nested_blocks[
(-nested_blocks['CVEGEO_parent'].isin(empty_blocks))&\
(-nested_blocks['CVEGEO_child'].isin(empty_blocks))
].groupby('CVEGEO_parent')['CVEGEO_child'].apply(
lambda children:[[child] for child in children]).to_dict()
print('# Parent Blocks', len(parent2children))

clusters_children = []
for i,idx_parent in enumerate(parent2children):
    
    if not i%100:
        print('# Parent', i)
        
    # Cluster children
    clusters_children += cluster_shapes(parent2children[idx_parent])
    
end = timer()
print('Computing Time:', round(end - start), 'sec')
```

# Agglomerative Clustering of Blocks By Admin Level

```python
start = timer()

print('Initialize each childless block in its own cluster')
ageb2blocks = pd.DataFrame(blocks.drop(
empty_blocks,errors='ignore').drop(
nested_blocks.CVEGEO_child.unique(),errors='ignore').reset_index().groupby(
['MUNIC','URBAN','AGEB'])['CVEGEO'].apply(lambda x:[[y] for y in x]))

print('Also list parents clusters')
ageb2childrenclusters = pd.DataFrame(pd.concat([
blocks.loc[cluster].reset_index().groupby(
['MUNIC','URBAN','AGEB'])['CVEGEO'].apply(list) 
for cluster in clusters_children]).groupby(
['MUNIC','URBAN','AGEB']).apply(list))

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Cluster blocks at the AGEB level')
start = timer()

print('Combine initial blocks and children clusters')
ageb2clusters = pd.DataFrame(index=ageb2blocks.index.union(ageb2childrenclusters.index))
ageb2clusters['CVEGEO'] = np.empty((len(ageb2clusters), 0)).tolist()
ageb2clusters.loc[ageb2blocks.index] = ageb2clusters.loc[ageb2blocks.index].add(ageb2blocks)
ageb2clusters.loc[ageb2childrenclusters.index] = \
ageb2clusters.loc[ageb2childrenclusters.index].add(ageb2childrenclusters)

# Randomize order
idx_ageb = ageb2clusters.sample(frac=1,random_state=0).index

with mp.Pool() as pool:
    clusters_ageb = pool.map(cluster_shapes, list(ageb2clusters.loc[idx_ageb,'CVEGEO']))

ageb2clusters.loc[idx_ageb,'CVEGEO'] = clusters_ageb.copy()

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Cluster blocks at the urban locality level')
start = timer()

urban2clusters = ageb2clusters.groupby(['MUNIC','URBAN']).sum()

# Randomize order
idx_urban = urban2clusters.sample(frac=1,random_state=0).index

with mp.Pool() as pool:
    clusters_urban = pool.map(cluster_shapes, list(urban2clusters.loc[idx_urban,'CVEGEO']))

urban2clusters.loc[idx_urban,'CVEGEO'] = clusters_urban.copy()

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Cluster blocks at the municipality level')
# Only one municipality contains small urban areas contiguous to one another
# The other small urban areas are in municipalities that do not contains any other urban areas
start = timer()

munic2clusters = urban2clusters.groupby(['MUNIC']).sum()

# Randomize order
idx_munic = munic2clusters.sample(frac=1,random_state=0).index

with mp.Pool() as pool:
    clusters_munic = pool.map(cluster_shapes, list(munic2clusters.loc[idx_munic,'CVEGEO']))

munic2clusters.loc[idx_munic,'CVEGEO'] = clusters_munic.copy()

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('List Clusters Features')
start = timer()

clusters_list = munic2clusters['CVEGEO'].sum()
pops_list     = [blocks.loc[cluster,'POB1'].sum() for cluster in clusters_list]
geoms_list    = [unary_union(blocks.loc[cluster,'geometry']) for cluster in clusters_list]
    
end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Extract Clusters Below the Threshold')
def get_below(clusters_list,pops_list,geoms_list,threshold=100):
    
    pos_below      = [pos_pop for pos_pop,pop in enumerate(pops_list) if pop<threshold]
    clusters_below = [cluster for pos_cluster,cluster in enumerate(clusters_list) if pos_cluster in pos_below]
    clusters_list  = [cluster for pos_cluster,cluster in enumerate(clusters_list) if pos_cluster not in pos_below]
    pops_list      = [pop for pos_pop,pop in enumerate(pops_list) if pos_pop not in pos_below]
    geoms_list     = [geom for pos_geom,geom in enumerate(geoms_list) if pos_geom not in pos_below]
    
    return clusters_below, clusters_list, pops_list, geoms_list

clusters_below, clusters_list, pops_list, geoms_list = get_below(clusters_list,pops_list,geoms_list)
```

```python
print('List Clusters Features')
start = timer()

clusters = gpd.GeoDataFrame(pd.DataFrame([clusters_list,pops_list],index=['CVEGEO','POB1']).T,geometry=geoms_list)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Dissolve by Urban Area')
start = timer()

urban = blocks[['URBAN', 'POB1', 'geometry']].dissolve(by='URBAN', aggfunc='sum')

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Find Closest Neighbor')
start = timer()

def neighbor_small_urban_areas(idx):
    return urban.iloc[np.argsort([geom.distance(urban.loc[idx,'geometry']) for geom in urban.geometry])[1:10]].index

small2neighbor = {}
for idx in small_urban:
    small2neighbor[idx] = neighbor_small_urban_areas(idx)
    
end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Import Municipalities')
start = timer()

path_to_shp = '../data/INEGI/'
munic = gpd.read_file(path_to_shp+'MunicipalALL.shp')
munic = munic.set_index('CVEGEO')[['POB1','geometry']].copy()
munic = munic.to_crs({'init': 'epsg:4326'})
print('# Municipalities:', munic.shape[0])

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Find Closest Munic')
start = timer()

def munic_small_urban_areas(idx):
    return munic.iloc[np.argmin([geom.distance(urban.loc[idx,'geometry']) for geom in munic.geometry])].name

small2munic = {}
for idx in small_urban:
    small2munic[idx] = munic_small_urban_areas(idx)
    
end = timer()
print('Computing Time:', round(end - start), 'sec')
```
