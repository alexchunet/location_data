- Population:

    - sum(POB1,Manzanas) = sum(POB1,AGEB) = sum(POB1,Urban) = 86984906
    - sum(POB1, Munic) = 112336538
    - POB1(AGEB_i) = sum_j(POB1(Manzanas_ij))
    - POB1(Urban_i) = sum_j(POB1(AGEB_ij))

- Shapes:
    - Manzanas do not overlay, except for the 10,005 which are nested (and 5 boundaries intersect)
    - 7 AGEB are also nested, but these are just mistakes
    - 2 Munic, 26 Urban, and 10,676 non-nested AGEB overlay at the boundary (more boundary intersect, as there are some differences between intersecting and overlaying...)
    - Municipalities intersections have an area of size zero
    - Sometimes Urban Area do not fully recover the underlying Manzanas
    - Sometimes the AGEB also do not fully recover the underlying Manzanas

```python
from timeit import default_timer as timer
import geopandas as gpd
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
print('Import Processed Census Data:')
start = timer()

path_to_shp = '../data/INEGI/'

manz  = gpd.read_file(path_to_shp+'ALLManzanas_SCINCE.shp')
ageb  = gpd.read_file(path_to_shp+'All_AgebUrbana.shp')
urban = gpd.read_file(path_to_shp+'loc_urb.shp')
munic = gpd.read_file(path_to_shp+'MunicipalALL.shp')

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Select Fields')

manz = manz[['CVEGEO','POB1','geometry']].copy()
ageb = ageb[['CVEGEO','POB1','geometry']].copy()
urban=urban[['CVEGEO','POB1','geometry']].copy()
munic=munic[['CVEGEO','POB1','geometry']].copy()

print('# Manzanas:', manz.shape[0])
print('# AGEB:', ageb.shape[0])
print('# Urban:', urban.shape[0])
print('# Municipalities:', munic.shape[0])
```

```python
print('Construct Missing Indices')

manz['AGEB']  = manz['CVEGEO'].apply(lambda x:x[:13])
manz['URBAN'] = manz['AGEB'].apply(lambda x:x[:9])
manz['MUNIC'] = manz['URBAN'].apply(lambda x:x[:5])

ageb['URBAN'] = ageb['CVEGEO'].apply(lambda x:x[:9])
ageb['MUNIC'] = ageb['URBAN'].apply(lambda x:x[:5])

urban['MUNIC'] = urban['CVEGEO'].apply(lambda x:x[:5])
```

```python
print('Reproject')
start = timer()

manz  = manz.to_crs({'init': 'epsg:4326'})
ageb  = ageb.to_crs({'init': 'epsg:4326'})
urban = urban.to_crs({'init': 'epsg:4326'})
munic = munic.to_crs({'init': 'epsg:4326'})
    
end = timer()
print('Computing Time:', round(end - start), 'sec')
```

# Check Nested Relationships

```python
print('Check Nested Relationships')
start = timer()

def nested_shapes(df,index_col='CVEGEO'):
    
    # Find Shapes that are Contained in one Another
    nested = gpd.sjoin(df,df,op='contains',lsuffix='parent',rsuffix='child')
    
    # Remove Original Shapes
    nested = nested[nested[(index_col+'_parent')]!=nested[(index_col+'_child')]].copy()
    
    return nested

manz_nested = nested_shapes(manz)
ageb_nested = nested_shapes(ageb)
urban_nested = nested_shapes(urban)
munic_nested = nested_shapes(munic)

print('Nested Manzanas:', manz_nested.shape[0])
print('Nested AGEB:', ageb_nested.shape[0])
print('Nested Urban:', urban_nested.shape[0])
print('Nested Municipalities:', munic_nested.shape[0])

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

# Check Overlay

```python
print('Check Overlay')
start = timer()

def intersected_shapes(df,index_col='CVEGEO'):
    
    # Intersect Shapes With Themselves
    intersect = gpd.overlay(df[[index_col,'geometry']], df[[index_col,'geometry']], how='intersection')
    
    # Remove Original Shapes
    intersect = intersect[(intersect[index_col+'_1']!=intersect[index_col+'_2'])].copy()
    
    return intersect

manz_intersected = intersected_shapes(manz)
ageb_intersected = intersected_shapes(ageb)
urban_intersected = intersected_shapes(urban)
munic_intersected = intersected_shapes(munic)

print('Intersected Manzanas:', manz_intersected.shape[0])
print('Intersected AGEB:', ageb_intersected.shape[0])
print('Intersected Urban:', urban_intersected.shape[0])
print('Intersected Municipalities:', munic_intersected.shape[0])
    
end = timer()
print('Computing Time:', round(end - start), 'sec')
```
