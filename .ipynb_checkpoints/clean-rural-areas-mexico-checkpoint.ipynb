```python
from timeit import default_timer as timer
import geopandas as gpd
import pandas as pd
from shapely.geometry import MultiPolygon, Polygon
import matplotlib.pyplot as plt
```

# Clean Shapes from Census

```python
path_to_data = '../data/'
```

```python
print('Import Shapes from Census')
start = timer()

ageb  = gpd.read_file(path_to_shp+'INEGI/All_AgebUrbana.shp')
urban = gpd.read_file(path_to_shp+'INEGI/loc_urb.shp')
munic = gpd.read_file(path_to_shp+'INEGI/MunicipalALL.shp')
print('# AGEB:', ageb.shape[0])
print('# Urban:', urban.shape[0])
print('# Municipalities:', munic.shape[0])

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Construct Missing Indices')

ageb['URBAN']  = ageb['CVEGEO'].apply(lambda x:x[:9])
ageb['MUNIC']  = ageb['URBAN'].apply(lambda x:x[:5])
urban['MUNIC'] = urban['CVEGEO'].apply(lambda x:x[:5])
```

# Aggregate Urban Shapes

```python
print('Import Census Blocks')
start = timer()

blocks = gpd.read_file(path_to_data+'blocks-mexico.geojson')
print('# Blocks:', blocks.shape[0])

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('% Valid BLOCKS:', blocks.geometry.is_valid.sum()/blocks.shape[0])
print('% Valid AGEB:', ageb.geometry.is_valid.sum()/ageb.shape[0])
print('% Valid URBAN:', urban.geometry.is_valid.sum()/urban.shape[0])
print('% Valid MUNIC:', munic.geometry.is_valid.sum()/munic.shape[0])
```

```python
print('Dissolve Blocks at the Urban Level')
start = timer()

urban_blocks = blocks[['URBAN', 'MUNIC', 'POB1', 'geometry']].dissolve(
by='URBAN', aggfunc={'MUNIC':'first','POB1':'sum'})

urban_blocks = urban_blocks.reset_index()[
['URBAN','MUNIC','POB1','geometry']
].sort_values(by=['URBAN']).reset_index(drop=True)

print('# Urban:', urban_blocks['URBAN'].unique().shape[0])
print('Pop:', urban_blocks['POB1'].sum())

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
urban_blocks.head()
```

```python
print('Dissolve AGEB at the Urban Level')
start = timer()

urban_ageb = ageb[['URBAN','MUNIC','POB1','geometry']].dissolve(
by=['URBAN'], aggfunc={'MUNIC':'first','POB1':'sum'})

urban_ageb = urban_ageb.reset_index()[
['URBAN','MUNIC','POB1','geometry']
].sort_values(by=['URBAN']).reset_index(drop=True)

print('# Urban:', urban_ageb['URBAN'].unique().shape[0])
print('Pop:', urban_ageb['POB1'].sum())

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
urban_ageb.head()
```

```python
print('Merge Urban Shapes')
start = timer()

urban_merged = pd.merge(pd.merge(
urban,
urban_ageb,
on=['URBAN','MUNIC','POB1'],
suffixes=['_urban','_ageb']),
urban_blocks).rename(
columns={'geometry':'geometry_blocks'})

print('# Urban:', urban_merged['URBAN'].unique().shape[0])
print('Pop:', urban_merged['POB1'].sum())

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Union Urban and AGEB Geometry')
start = timer()

urban_merged['geometry_urban_ageb'] = urban_merged.apply(lambda x:x.geometry_urban.union(x.geometry_ageb),1)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Extract shapes from union of blocks that is neither in AGEB nor in urban shape')
start = timer()

urban_merged['geometry_diff'] = urban_merged.apply(lambda x:x.geometry_blocks.difference(x.geometry_urban_ageb),1)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('Max. Missing Area:', urban_merged['geometry_diff'].apply(lambda x:x.area).max())
```

```python
print('Union of all Urban Geometries')
start = timer()

urban_merged['geometry_union'] = urban_merged.apply(lambda x:x.geometry_urban_ageb.union(x.geometry_diff),1)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
def extract_poly_coords(geom):
    if geom.type == 'Polygon':
        exterior_coords = geom.exterior.coords[:]
        interior_coords = []
        for interior in geom.interiors:
            interior_coords += interior.coords[:]
    elif geom.type == 'MultiPolygon':
        exterior_coords = []
        interior_coords = []
        for part in geom:
            epc = extract_poly_coords(part)  # Recursive call
            exterior_coords += epc['exterior_coords']
            interior_coords += epc['interior_coords']
    else:
        raise ValueError('Unhandled geometry type: ' + repr(geom.type))
    return {'exterior_coords': exterior_coords,
            'interior_coords': interior_coords}
```

```python
print('Convert Interior Rings Into Polygons')

def get_interiors(x):
    # Works with two levels only
    if x.type=='Polygon':
        interiors = []
        for interior in x.interiors:
            interiors.append(interior)
    elif x.type=='MultiPolygon':
        interiors = []
        for y in x:
            for interior in y.interiors:
                interiors.append(interior)
    return interiors

urban_merged['interiors'] = urban_merged.geometry_union.apply(get_interiors)
urban_merged['geometry_interiors'] = urban_merged['interiors'].apply(lambda x:MultiPolygon([Polygon(y) for y in x]))
```

```python
print('Union Urban Area and Inner Holes')
start = timer()

urban_merged['geometry'] = urban_merged.apply(lambda x:x.geometry_union.union(x.geometry_interiors),1)

urban_merged = gpd.GeoDataFrame(urban_merged,
geometry=urban_merged.geometry).sort_values(by=['URBAN']).reset_index(drop=True)

urban_merged.POB1 = urban_merged.POB1.astype(int)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
def plot_urban(idx):
    fig,ax=plt.subplots(figsize=(10,10))
    # rural[rural['MUNIC']==urban.loc[idx,'MUNIC']].plot(ax=ax,alpha=0.1,facecolor='y',edgecolor='k')
    urban_merged.loc[[idx]].plot(ax=ax,alpha=0.1,facecolor='m',edgecolor='k')
    urban.loc[[idx]].plot(ax=ax,alpha=0.1,facecolor='r',edgecolor='k')
    urban_ageb.loc[[idx]].plot(ax=ax,alpha=0.1,facecolor='g',edgecolor='k')
    urban_blocks.loc[[idx]].plot(ax=ax,alpha=0.5,facecolor='b',edgecolor='k')
    plt.axis('off')
```

```python
for idx in urban_merged.apply(lambda x:x['geometry_blocks'].area/x['geometry'].area,1).sort_values().index[:10]:
    plot_urban(idx)
```

# Extract Rural Areas

```python
start = timer()

print('Extract Rural Areas')
rural = gpd.overlay(munic, urban_merged[['geometry']], how='difference')

print('Compute Rural Population')
rural = pd.merge(rural,urban.groupby('MUNIC')['POB1'].sum().reset_index(),on='MUNIC')
rural['POB1'] = rural['POB1_x'] - rural['POB1_y']

rural = rural[['MUNIC','POB1','geometry']].sort_values(by='MUNIC').reset_index(drop=True)
rural.POB1 = rural.POB1.astype(int)

print('Create homogenous geometry for exporting')
rural['geometry']=[MultiPolygon([feature]) if type(feature)==Polygon else feature for feature in rural['geometry']]

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('# Rural Areas', rural.shape[0])
print('# Rural Munic', rural.MUNIC.unique().shape[0])
print('Pop:', rural.POB1.sum())
print('% Valid RURAL:', rural.geometry.is_valid.sum()/rural.shape[0])
```

# Save

```python
print('Save Rural Areas')
start = timer()

rural.to_file(path_to_data+"rural-mexico.geojson", driver='GeoJSON')

end = timer()
print('Computing Time:', round(end - start), 'sec')
```
